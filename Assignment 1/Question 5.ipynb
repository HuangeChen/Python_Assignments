Use Gutenberg and Web_text data. Find out what are the top 5 words that Shakespeare used but we do not use in currently. Take top 50 words from Shakespeare (all 3 books) and top 50 from Web_text (all the records). Remove punctuation and stop words. Remove the words we still use today, and get the unused list. Show the top 5 elements.

```python
import nltk
```

```python
from nltk.corpus import gutenberg
```

```python
nltk.corpus.gutenberg.fileids()
```

```python
words1=gutenberg.words('shakespeare-caesar.txt')
words2=gutenberg.words('shakespeare-hamlet.txt')
words3=gutenberg.words('shakespeare-macbeth.txt')
total_words=words1+words2+words3
print(total_words)
```

```python
fdist1= nltk.FreqDist(total_words);
```

```python
vocabulary1=fdist1.keys()
```

```python
print(vocabulary1)
```

```python
from collections import Counter
```

```python
word_counts=Counter(vocabulary1)
```

```python
top_fifty=word_counts.most_common(50)
```

```python
print(top_fifty)
```

```python
def unusual_words(text):
    text_vocab=set(w.lower() for w in text if w.isalpha())
    english_vocab=set(w.lower() for w in nltk.corpus.words.words())
    unusual=text_vocab.difference(english_vocab)
    return sorted(unusual)


```

```python
unused1=unusual_words(nltk.corpus.gutenberg.words( 'shakespeare-caesar.txt'))
```

```python
unused2=unusual_words(nltk.corpus.gutenberg.words( 'shakespeare-hamlet.txt'))
```

```python
unused3=unusual_words(nltk.corpus.gutenberg.words( 'shakespeare-macbeth.txt'))
```

```python
unused_word=unused1+unused2+unused3
```

```python
print(unused_word)
```
